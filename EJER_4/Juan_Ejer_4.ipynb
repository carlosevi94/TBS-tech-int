Abro el dataset poniendo el numero de canción como index.

```python
import pandas as pd
df = pd.read_csv('data/train.csv').set_index('Number')
df
```

De estos 386 registros, tenemos que sacar cuantas canciones por artista hay:

```python
df_artist = df[['artist']].copy()
conteo = pd.DataFrame(df_artist.groupby(['artist'])['artist'].count().sort_values(ascending=False))
conteo.head(10)
```

Saco como query SQL:

```python
SELECT
    artist AS ARTISTA,
    COUNT(artist) AS CONTEO
FROM
    spotify_dataset
GROUP BY artist
SORT BY 2 DESC
```

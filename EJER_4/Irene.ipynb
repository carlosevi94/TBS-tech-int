genero_especifico : Variable booleana que comprueba si el genero de la canción es más especifico (tiene más de una palabra).
Ej: Rock es un grupo general
Rock metal es un grupo específico
Rap es un grupo general
rap urban es un grupo específico

```python
import pandas as pd
df = pd.read_csv("train.csv")
df
```

```python
#primero convertimos la columna género a string

df["top genre"] = df["top genre"].astype(str)

#y volvemos a comprobar los tipos

df.dtypes

```

```python
#contamos el numero de caracteres que tiene la columna
df["palabras"] = df["top genre"].str.split()
df
```

```python
#CREAMOS LA COLUMNA CONTADOR QUE CUENTA LA LONGITUD DE PALABRAS
df["contador"] = df["palabras"].str.len()
df
```

```python
#CREAMOS LA COLUMNA BOOLEANA, SI LA LONGITUD ES MAYOR DE 2, FALSE, SINO TRUE
# ESPECIFICO TRUE, GENERAL FALSE

df["bool"] = df.contador < 2
df
```
